```python
df["Order Date"] = pd.to_datetime(df["Order Date"])
df["Ship Date"] = pd.to_datetime(df["Ship Date"])
```

```python
df.info()
```

```python
total_revenue = df["Sales"].sum()
total_profit = df["Profit"].sum()
total_orders = df["Order ID"].nunique()
total_customers = df["Customer Name"].nunique()
average_discount = df["Discount"].mean()

print("Total Revenue:", round(total_revenue, 2))
print("Total Profit:", round(total_profit, 2))
print("Total Orders:", total_orders)
print("Total Customers:", total_customers)
print("Average Discount:", round(average_discount, 2))
```

```python
region_summary = df.groupby("Region")[["Sales", "Profit"]].sum().sort_values(by="Sales", ascending=False)
region_summary
```

```python
category_summary = df.groupby("Category")[["Sales", "Profit"]].sum().sort_values(by="Profit", ascending=False)
category_summary
```

```python
subcategory_summary = df.groupby("Sub-Category")[["Sales", "Profit"]].sum().sort_values(by="Profit", ascending=False)
subcategory_summary
```

```python
top_customers = df.groupby("Customer Name")[["Sales", "Profit"]].sum().sort_values(by="Sales", ascending=False).head(10)
top_customers
```

```python
monthly_sales = df.groupby(df["Order Date"].dt.to_period("M"))["Sales"].sum()
monthly_sales
```

```python
monthly_sales.index = monthly_sales.index.astype(str)
```

```python
df["Shipping Days"] = (df["Ship Date"] - df["Order Date"]).dt.days
ship_mode_summary = df.groupby("Ship Mode")["Shipping Days"].mean().sort_values()
ship_mode_summary
```

```python
discount_summary = df.groupby("Discount")[["Sales", "Profit"]].sum().sort_index()
discount_summary
```

```python
region_summary["Sales"].plot(kind="bar")
plt.title("Sales by Region")
plt.xlabel("Region")
plt.ylabel("Total Sales")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
category_summary["Profit"].plot(kind="bar")
plt.title("Profit by Category")
plt.xlabel("Category")
plt.ylabel("Total Profit")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
top_customers["Sales"].sort_values().plot(kind="barh")
plt.title("Top 10 Customers by Sales")
plt.xlabel("Total Sales")
plt.ylabel("Customer Name")
plt.tight_layout()
plt.show()
```

```python
monthly_sales.plot(kind="line")
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
discount_summary["Profit"].plot(kind="bar")
plt.title("Profit by Discount Level")
plt.xlabel("Discount")
plt.ylabel("Total Profit")
plt.tight_layout()
plt.show()
```
